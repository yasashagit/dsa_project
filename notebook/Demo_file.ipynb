#### Array Data Structure:

**Array**
An Array is a data structure that stores elements in contiguous memory locations (next to each other in one continuous block of memory, without gaps), allowing fast access to any element using its index. Arrays are fixed-order collections, meaning the position of each element is predetermined by its index. They are ideal for situations where you need quick, random access to elements.

<br>

Process and Characteristics:

1) Elements are stored next to each other in memory.
2) Each element can be accessed directly using its index (e.g., array[3]).
3) A real low level array stores elements of only one data type.
4) Inserting an element at the end is fast, but inserting in the middle may require shifting other elements.
5) Deleting an element also requires shifting elements to fill the gap.
6) Searching requires scanning the array unless the array is sorted.

<br>

**Complexity**: Fast access ($O(1)$), slow insertions/deletions in the middle ($O(n)$)

#### List Data Structure:

**List**

A List in Python is a dynamic, ordered collection of elements. Unlike arrays, Python lists can grow or shrink at runtime because they are implemented as dynamic arrays under the hood. Lists allow mixed data types, support indexing, and offer many built-in operations for adding, removing, and searching elements.

<br>

Process and Characteristics:

1) Lists maintain the order of elements as they were inserted.
2) Elements can be accessed directly using their index (e.g., lst[2]).
3) Appending elements at the end is efficient because Python over-allocates memory.
4) Inserting or deleting elements in the middle requires shifting other elements.
5) Searching for a value requires scanning the list unless the list is sorted.

<br>

**Complexity**: Fast access: ($O(1)$), insert/delete in the middle: ($O(n)$), search: ($O(n)$)

#### Linked List Data Structure

**Linked List**

A Linked List is a linear data structure where each element (node) stores a value and a pointer/reference to the next node. Nodes are not stored in contiguous memory, allowing fast insertions and deletions at specific positions.

<br>

Process and Characteristics:

1) Each node contains a value and a link to the next node.
2) Access requires starting from the head and following links, no direct indexing.
3) Insertion at the beginning is extremely fast (no shifting needed).
4) Deletion of a node requires knowing the previous node.
5) Useful for dynamic data where frequent insertions/deletions occur.

<br>

**Complexity**: Access: ($O(n)$), insertion at front: ($O(1)$), deletion (with previous node known): ($O(1)$), search: ($O(n)$)

#### Stack Data Structure

**Stack**

A Stack is a Last-In-First-Out (LIFO) data structure. The most recently added element is the first one to be removed. Stacks are used for undo systems, recursion, compiler parsing, and backtracking.

<br>

Process and Characteristics:

1) Elements are added using push() and removed using pop().
2) Only the top element is accessible — no random access.
3) Common operations include push, pop, and peek.
4) Ideal for managing nested tasks or reversing sequences.

<br>

**Complexity**: Push: ($O(1)$), Pop: ($O(1)$), Peek: ($O(1)$), Search: ($O(n)$)

```python
# Creating the stack

language_stack = Stack()

# Pushing elements onto the stack (O(1))

print("Pushing items onto the stack...")
for lang in languages:
    language_stack.push(lang)

print("Stack after pushing all items:")
print(language_stack, "\n")

# Peeking at the top element (O(1))

print("Top of the stack (peek):")
print(language_stack.peek(), "\n")

# Popping elements (O(1))

print("Popping two items from the stack...")
popped1 = language_stack.pop()
popped2 = language_stack.pop()

print("Popped:", popped1)
print("Popped:", popped2)
print("Stack after popping:")
print(language_stack, "\n")

# Checking if stack is empty (O(1))

print("Is the stack empty?")
print(language_stack.is_empty())
```

#### Queue Data Structure

**Queue**

A Queue is a First-In-First-Out (FIFO) data structure. The first element added is the first to be removed. Queues are used for scheduling, buffering, simulations, and breadth-first search.

<br>

Process and Characteristics:

1) Elements enter from the rear and leave from the front.
2) Main operations: enqueue() (add), dequeue() (remove).
3) Only the front element can be accessed at any time.
4) Suitable for tasks that must be processed in order.

<br>

Complexity: Enqueue: ($O(1)$), Dequeue: ($O(1)$), Peek front/rear: ($O(1)$), Search: ($O(n)$)

```python
Image("../images/queue.gif", width=500)
```

```python
# Using the same dataset one last time

languages = [
    ("Python", 3.1),
    ("Java", 1.4),
    ("JavaScript", 1.9),
    ("C++", 0.5),
    ("C#", 1.6),
    ("TypeScript", 2.1),
    ("SQL", 4.0),
    ("C", 0.4),
    ("Go", 0.9),
    ("PHP", 3.4)
]
```

#### Sorting Algorithms:

**Bubble sort**
<br>
Bubble Sort repeatedly steps through the list, compares adjacent elements, and swaps them if they are in the wrong order. The process is repeated until the list is sorted. With each pass, the largest unsorted element "bubbles up" to its correct position at the end of the list. 

<br>

Process:
1) Start at the beginning of the list.
2) Compare the first two elements. If the first is larger than the second, swap them.
3) Move one position forward and repeat the comparison and swap for the next pair.
4) Repeat this pass until the end of the list is reached.
5) The largest element is now in its final sorted position.
6) Repeat the entire process for the remaining unsorted portion of the list.

<br>

**Complexity:** $O(n^2)$ (slow for large lists).

**Insertion Sort**
<br>
Insertion Sort builds the final sorted array one item at a time. It iterates through the input elements and inserts each element into its correct position within the already sorted portion of the array.

<br>

Process:
1) The first element is considered the initial sorted subarray (of size 1).
2) Take the next element (the first in the unsorted portion). 
3) Compare the new element with elements in the sorted subarray, moving backward.
4) Shift all elements in the sorted subarray that are greater than the current one position to the right to create space.
5) Insert the element into the gap created.
6) Repeat the process until all elements are in the sorted subarray.
<br>

**Complexity:** $O(n^2)$ (comparisons & swaps).

**Merge Sort**
<br>
Merge Sort is a highly efficient algorithm. It works by breaking down a list into two sub-lists until each sub-list contains only one element, and then repeatedly merging the sub-lists back together in a sorted position.  

Process:
1) Divide: Recursively split the list into two halves until the sub-lists contain a single element (a single-element list is inherently sorted).
2) Merge: Merge the sub-lists back together by comparing the elements in the two lists. Always take the smaller element from the front of the two sub-lists and place it into the combined list.
3) Repeat the merge step until the entire list is reassembled into a single, sorted list.

**Complexity:** $O(n \log n)$ (fast for large lists).

<img src="https://upload.wikimedia.org/wikipedia/commons/c/cc/Merge-sort-example-300px.gif" style="width: 50%;">

```python
customers_id = [order['customer_id']for order in orders]
print(customers_id)
```

```python
sorted_customers_id = merge_sort(customers_id)
print(sorted_customers_id)
```

```python
input_sizes = [100, 200, 400, 600, 800, 1000, 1500, 2000]
bubble_times = []
insertion_times = []
merge_times = []

for size in input_sizes:
    data = [random.randint(0, 10000) for _ in range(size)]
    
    data_copy = data.copy()
    start = time.perf_counter()
    bubble_sort(data_copy)
    end = time.perf_counter()
    bubble_times.append(end - start)
    
    data_copy = data.copy()
    start = time.perf_counter()
    insertion_sort(data_copy)
    end = time.perf_counter()
    insertion_times.append(end - start)
    
    data_copy = data.copy()
    start = time.perf_counter()
    merge_sort(data_copy)
    end = time.perf_counter()
    merge_times.append(end - start)

plt.figure(figsize=(10, 6))
plt.plot(input_sizes, bubble_times, label='Bubble Sort $O(n^2)$', marker='o', color='red')
plt.plot(input_sizes, insertion_times, label='Insertion Sort $O(n^2)$', marker='s', color='orange')
plt.plot(input_sizes, merge_times, label='Merge Sort $O(n \log n)$', marker='^', color='green')

plt.title('Sorting Algorithms Performance Comparison')
plt.xlabel('Input Size (Number of Elements)')
plt.ylabel('Time Taken (Seconds)')
plt.legend()
plt.grid(True)
plt.show()
```

### 3) Trees

#### Binary Tree

A Binary Tree is a hierarchical data structure where each node can have up to two children: a left child and a right child. It is used to represent relationships, hierarchical structures, and forms the foundation for more advanced trees such as Binary Search Trees and Balanced Trees.

<br>

Process and Characteristics:

1) Each node contains a value and pointers/references to its left and right child.
2) The tree begins at the root node, from which all other nodes branch.
3) Nodes without children are called leaf nodes.
4) Traversal can be performed in multiple ways:
5) In-order (Left → Root → Right)
6) Pre-order (Root → Left → Right)
7) Post-order (Left → Right → Root)
8) Level-order (Breadth-first)
9) Binary Trees do not enforce any ordering of values.
<br>
Complexity: Traversal: $O(n)$, Search/Insert/Delete: $O(n)$ in the worst case (unstructured tree)

#### Binary Search Tree (BST)

A Binary Search Tree is a Binary Tree with a special ordering rule:

- Left subtree contains values less than the node

- Right subtree contains values greater than the node

- This ordering makes searching efficient.

  <br>

Process and Characteristics:

1) Insert a value by comparing it to the root:
2) Smaller → go left
3) Larger → go right
4) Repeat the comparison until you find an empty spot.
5) Searching follows the same logic — each comparison eliminates half the tree.
6) In-order traversal of a BST produces a sorted list of values.
7) If the tree becomes “unbalanced,” performance can degrade.

<br>

Complexity: Best/Average: Search/Insert/Delete → $O(\log n)$, Worst case (unbalanced): $O(n)$

```python
Image("../images/BST.gif", width=500)
```

```python
# Creating the Binary Search Tree (BST)

bst = BST()

# Inserting elements into the BST and we get average O(log n) and  worst O(n)

print("Inserting items into the BST...\n")
for lang in languages:
    bst.add(lang)

print("In-order traversal (sorted by language name):")
bst.inprint()
print("\n")

print("Pre-order traversal (Root → Left → Right):")
bst.preprint()
print("\n")

print("Post-order traversal (Left → Right → Root):")
bst.postprint()
print("\n")

print("Level-order traversal (Breadth-first by levels):")
bst.levelprint()
print("\n")

# Searching for elements and we have average O(log n) and worst case O(n)

print("Searching for ('Python', 3.1):")
bst.search(("Python", 3.1))

print("Searching for ('Rust', 1.2):")
bst.search(("Rust", 1.2))
print("\n")

# Deleting a node

print("Deleting ('C#', 1.6) from the BST...")
bst.root = bst.delete(bst.root, ("C#", 1.6))

print("In-order traversal after deletion:")
bst.inprint()
print("\n")

# Visualizing the BST

bst.visualize()
```

#### Balanced Tree

A Balanced Tree is any Binary Tree (including BSTs) that maintains its height close to minimal after insertions and deletions.
Examples include: AVL Trees, Red-Black Trees, B-Trees, and 2-3 Trees.

The goal is to ensure performance stays predictable and fast.

<br>

Process and Characteristics:

1) The tree monitors its height or balance factor after each insertion/deletion.
2) If the tree becomes too heavy on one side, it performs rotations to restore balance.
3) Height is kept around $O(\log n)$, ensuring fast searching.
4) Balanced Trees guarantee efficient operations even with random input.
5) They are widely used inside databases, index structures, and language runtimes.

<br>

Complexity: Search: $O(\log n)$, insert: $O(\log n)$, delete: $O(\log n)$ (Always guaranteed, unlike a normal BST.)

### 4) Hash Tables

**Hash Table** (or Hash Map) is a highly efficient data structure used for storing and retrieving data quickly. Unlike lists or arrays, a Hash Table uses a specialized function to locate data instantly.
  
Structure and Mechanism
- Key-Value Storage: A Hash Table stores data as key-value pairs (like a dictionary in Python).
- Hash Function: When a key is inserted, a hash function is used to convert that key (e.g., a customer name or ID) into a small integer, which serves as the index or "bucket" in an underlying array.
- Direct Access: This calculated index allows the structure to jump directly to the data's location, leading to near-instantaneous retrieval.

Time Complexity and Collisions. 
- Average Case Time Complexity: $O(1)$ (Constant Time). This is the key advantage. Whether it 10 items or 10 million items, the lookup time generally stays the same because it involves one calculation (hashing) and one direct array access.
- Worst Case Time Complexity: $O(n)$. This occurs if many different keys map to the same index, causing a collision.
- Collision Handling: Collisions are usually handled by chaining, it means if a spot is already taken, it starts a list there and attach the new item to the end of it.

```python
input_sizes = [10000, 50000, 100000, 500000, 1000000]

TABLE_SIZE = 2**20 

linear_times = []
hash_times = []

for size in input_sizes:
    data_list = [{'id': i, 'value': f'data_{i}'} for i in range(size)]
    
    hash_t = HashTable(size=TABLE_SIZE)
    
    for i in range(size):
        hash_t.add(i, f'data_{i}')
        
    target = size - 1 
    
    start = time.perf_counter()
    linear_search(data_list, target)
    end = time.perf_counter()
    linear_times.append(end - start)
    
    start = time.perf_counter()
    hash_t.search(target) 
    end = time.perf_counter()
    hash_times.append(end - start)

plt.figure(figsize=(10, 6))

plt.plot(input_sizes, linear_times, label='Linear Search O(n)', color='red', marker='o')
plt.plot(input_sizes, hash_times, label='Hash Table Search O(1)', color='blue', marker='x')

plt.title('Performance Comparison: Hash Table Search vs Linear Search ')
plt.xlabel('Input Size (N)')
plt.ylabel('Time Taken (Seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
